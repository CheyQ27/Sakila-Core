```python
import pymysql
pymysql.install_as_MySQLdb()
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
connection = "mysql+pymysql://root:root@localhost/sakila"

```

```python
engine = create_engine(connection)
```

```python
engine
```

```python
q1 ="""SELECT customer.first_name, customer.last_name, 
    customer.email, address.address, city.city
FROM customer
JOIN address
JOIN city
WHERE customer.address_id=address.address_id
AND city.city_id=address.city_id
AND city.city_id='312';"""
pd.read_sql(q1,engine)
```

```python
q2 ="""select film.title, film.description, film.release_year,
    film.rating, film.special_features
from film 
join category
join film_category
where film.film_id=film_category.film_id
and film_category.category_id=film_category.category_id
and category.name='Comedy';"""
pd.read_sql(q2, engine)
```

```python
q3 ="""select actor.last_name, film.title, film.release_year
from film
join actor
join film_actor
where film.film_id=film_actor.film_id
and film_actor.actor_id=actor.actor_id
and actor.last_name='LOLLOBRIGIDA';"""
pd.read_sql(q3, engine)
```

```python
q4 ="""select actor.last_name, actor.first_name
from film
join actor
join film_actor
where film.film_id=film_actor.film_id
and film_actor.actor_id=actor.actor_id
and film.title='BINGO TALENTED'"""
pd.read_sql(q4,engine)
```

```python
q5 ="""select avg(payment.amount) as amt, payment.customer_id
from payment 
where payment.amount > 2*(select avg(payment.amount) from payment);"""
pd.read_sql(q5, engine)
```
